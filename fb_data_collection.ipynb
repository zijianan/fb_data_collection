```python
for folder in os.listdir('data'):
    if "file.csv" in os.listdir('data/'+folder):
        current_csv = pd.read_csv('data/'+folder+'/file.csv').dropna(subset=['TIME'])
        current_csv = current_csv.dropna(subset=['TYPE'])
        current_csv[['LINKS']] = current_csv[['LINKS']].fillna('')
        hashtag = current_csv[current_csv['TYPE'].str.contains('#')]
        for index,row in tqdm(hashtag.iterrows(),total = hashtag.shape[0]):
            splitdata = row['TYPE'].split()
            for sub in splitdata:
                if "#" in sub:
                    current_csv.at[index,'LINKS'] += sub+' '
        
        current_csv = current_csv.rename(columns={"TYPE": "POST", "TITLE": "REPLY", "STATUS": "LINKS", "LINKS": "HASHTAG"})
        if folder in page_list['foldername'].values:
            filename = page_list.loc[page_list['foldername'] == folder]['name'].values[0].replace(' ','_')
            filename = filename.replace('/','_')
            current_csv.to_csv('cleaned/'+filename+'.csv',index=False)
        else:
            try:
                prename = dictname[folder]
                filename = prename.replace(' ','_')
                filename = filename.replace('/','_')
                current_csv.to_csv('cleaned/'+filename+'.csv',index=False)
            except:
                print(folder)
    else:
        print('dont have csv',folder)
```
